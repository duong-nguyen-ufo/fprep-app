```python
import os 
import openai
from openai import OpenAI
import getpass
import pandas as pd
```

```python
# Authenticate OpenAI API
key = getpass.getpass('Paste your OpenAI API key: ')
os.environ['OPENAI_API_KEY'] = key
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
# Get user's name
user_name = input('What is your name?')
```

```python
# User's kitchen setup 
stove_burner = input('How many stove burners do you have?')
oven_rack = input('How many oven racks do you have?')
sous_vide_bag = input('How many sous vide bags do you have?')
large_pan = input('How many large size pans do you have?')
medium_pan = input('How many medium size pans do you have?')
small_pan = input('How many small size pans do you have?')
large_pot = input('How many large size pots do you have?')
medium_pot = input('How many medium size pots do you have?')
small_pot = input('How many small size pots do you have?')
food_processor = input('How many food processors do you have?')
blender_cup = input('How many blender cups do you have?')
crock_pot = input('How many crock pots do you have?')
rice_cooker = input('How many rice cookers do you have?')
thermometer = input('How many thermometers do you have?')
```

```python
# Append new user profile to store in dataframe
df = pd.DataFrame(columns=['user_name', 'stove_burner', 'oven_rack', 'sous_vide_bag', 'large_pan', 'medium_pan', 
                          'small_pan', 'large_pot', 'medium_pot', 'small_pot', 'food_processor', 'blender_cup', 'crock_pot', 'thermometer'])
user_profile = [user_name, stove_burner, oven_rack, sous_vide_bag, large_pan, medium_pan, small_pan, large_pot, medium_pot, small_pot, food_processor, blender_cup, crock_pot, thermometer]

user_profile_series = pd.Series(dict(zip(df.columns, user_profile)))
df = df._append(user_profile_series, ignore_index=True)              
df
```

```python
# User's food and cooking preferences

style = input('What style do you want your recipes to be? Simple and minimal / Complex with layers of flavors / Balance - not too simple but not too complex')
calories = input('How many calories do you want for the total of all three meals per day?')
additional_preference = input('Any additional preference? Such as the type of cuisine, the use of ingredients or cooking methods?')
temp_unit = input('What is your temperature measure unit?')
liquid_unit = input('What is your liquid measure unit?')
mass_unit = input('What is your mass measure unit?')
```

```python
# User's needs for recipes 
plan_name = input("Give this meal plan a name, such as Week April 21-26, so that it's easier for you to identify the meal plan")
existing_ingredients = input('Are there any ingredients that you already have and want to incoroporate into the recipes if possible?')
days = input('How many days do you want to cook for?')
breakfast = input('What do you want to have for breakfast?')
lunch = input('What do you want to have for lunch?')
dinner = input('What do you want to have for dinner?')
```

```python
# Call OpenAI API
client = OpenAI()

# System prompt and model
system_prompt = '''You are an experienced chef. You have many experiences in cooking healthy dishes and cooking multiple dishes simultaneously in a very efficient, methodological way.'''
model = 'gpt-4o-mini'
```

```python
# Generate recipes and grocery list prompt

recipe_user_prompt = f'''
Please help write a meal plan and grocery list tailored to my needs.

Step 1: Understand all the information that you need to write a cooking plan for my  needs.
- My style of cooking is {style}.
- I want to cook for {days} for the next week.
- I need all meals combined to have {calories} kcal per day.
- For breakfast, I want {breakfast}. 
- For lunch, I want {lunch}.
- For dinner, I want {dinner}.
- My preference for the food is {additional_preference}.
- The existing ingredients that I want to incorporate into the recipes, if possible, are {existing_ingredients}.
- My measure units for temperature, liquid, and mass, respectively, are: {temp_unit}, {liquid_unit}, {mass_unit}
- For my cooking equipment, besides the basic equipment like knives, spatulas, spoons, forks, mixing bowls, measuring cups, etc., here is the list of the key equipment I have:
  + {stove_burner} stove burners 
  + {oven_rack} oven racks 
  + {sous_vide_bag} sous vide bags 
  + {large_pan} large pans 
  + {medium_pan} medium pans
  + {small_pan} small pans
  + {large_pot} large pots
  + {medium_pot} medium pots
  + {small_pot} small pots
  + {food_processor} food processors
  + {blender_cup} blender cups
  + {crock_pot} crock pots
  + {rice_cooker} rice cookers
  + {thermometer} thermometers

Step 2: Generate the recipes for each meal given in Step 1. For each meal: 
- Give the meal a name
- State the cooking method
- Estimate the calories for this meal for one day
- Write the recipe. Ensure that the recipe meets the needs in Step 1 and use the cooking equipment only listed in Step 1. 
  -- Ingredients: List the ingredients. All ingredients must have the amount needed for the recipe. 
  -- Equipment: List the equipment needed for the recipe
  -- Instructions: List the steps to cook the recipe

Step 3: Combine the ingredient lists from the three meals into one list. This will be my grocery list. 
- This grocery list is split into two sections: 
  -- 1. Ingredients I already have
  -- 2. Ingredients I need to buy.  All ingredients must have the amount needed for the recipes. 
- For each ingredient, state the amount needed for that ingredient. If two recipes share the same ingredient, list how much of that ingredient is needed for each meal. For example: "Mixed greens: 5 cups for breakfast, 5 cups for dinner". 

Step 4: Export the output in the format below. Ensure that the output does not include any XML tags. 
Recipes for {plan_name}
**Recipes**
Breakfast: (recipe in Step 2 for {breakfast})
Lunch: (recipe in Step 2 for {lunch})
Dinner: (recipe in Step 2 for {dinner})

**Grocery list**
(summarized list of ingredients in Step 3)
'''
```

```python
recipe_response = client.chat.completions.create(
    model = model,
    messages = [
    {'role': 'system', 'content': system_prompt},
    {'role': 'user', 'content': recipe_user_prompt}
    ],
    temperature = 0.1
).choices[0].message.content

print(recipe_response)
```

```python
# Ask user if they want to update the recipe 
want_to_update = input('Do you want to update the recipes?')

if want_to_update.lower() == 'no':
    recipe_response = recipe_response
    print("Cool! Let's move on to the cooking plan.")
    
if want_to_update.lower() == 'yes':
    update_questions = list()
    update_bot_responses = list()
    update_messages = list()

    update_messages.append({'role': 'system', 'content': system_prompt})
    update_messages.append({'role': 'assistant', 'content': recipe_response + 
                                    f'''Output format: Keep the current format of the {recipe_response} but indicate which part of the {recipe_response} was updated'''})

    while True: # create an infinite loop of dialog until user exits
        current_question = input('Which parts of the recipes do you want to adjust?')
        
        if current_question.lower() in ['exit', 'quit', 'nothing else']:
            print("Cool! Let's move on to the cooking plan.")
            recipe_response = update_bot_responses[-1]
            # time.sleep(2) # wait for two seconds before ending so it's better UX
            break

        if current_question == '': #if user just hits Enter without a question, the conversation still continues
            continue 

        update_messages.append({'role': 'user', 'content': current_question})
        update_questions.append(current_question)
    
        recipe_update_response = client.chat.completions.create(
                                    model = model,
                                    messages = update_messages,
                                    temperature = 0.1
        )
      
        current_response = recipe_update_response.choices[0].message.content
        print(f'{current_response}')

        update_bot_responses.append(current_response)
        update_messages.append({'role': 'assistant', 'content': current_response + f'''Output format: Keep the current format of the {recipe_response} but indicate which part of the {recipe_response} was updated'''})
        
        print('\n' + '-'*50 + '\n')  
```

```python
# Generate cooking steps

cooking_user_prompt = f'''
Please help write a cooking plan, step by step, that is tailored to my needs below.

Step 1: Understand all the information that you need to write a cooking plan for my  needs.
- I have very limited time, so I want to cook all three meals for all {days} days at once, then store the meals for the week. 
- The recipes I want to cook are the section **Recipes** in {recipe_response}
- For my cooking equipment, besides the basic equipment like knives, spatulas, spoons, forks, mixing bowls, measuring cups, etc., here is the list of the key equipment I have:
  + {stove_burner} stove burners 
  + {oven_rack} oven racks 
  + {sous_vide_bag} sous vide bags 
  + {large_pan} large pans 
  + {medium_pan} medium pans
  + {small_pan} small pans
  + {large_pot} large pots
  + {medium_pot} medium pots
  + {small_pot} small pots
  + {food_processor} food processors
  + {blender_cup} blender cups
  + {crock_pot} crock pots
  + {thermometer} themometers

Step 2: Generate the cooking plan, step by step. The goal of the cooking plan is to make the cooking as efficient as possible with little waiting time.
The plan should meet the following criteria:
- Have very minimal in-cooking equipment washing time
- Utilize all the equipment as much as possible to prepare or cook multiple ingredients simultaneously. 
- Include the steps and time for washing produce and washing equipment, if any.
- Estimate the time taken for each step
- Because the food is cooked and stored for the week ahead, include the storing and packing step. 

The cooking plan should start from the minute 0 as the starting point of the cooking timeline, then add each step with the step's duration. 

Step 3: Export the output in the format below. Ensure that the output does not include any XML tags. 
Cooking plan for {plan_name}
**Total time**: (estimated total time)
**Steps**
- (cooking step)
'''
```

```python
cooking_response = client.chat.completions.create(
        model = model,
        messages = [
        {'role': 'system', 'content': system_prompt},
        {'role': 'user', 'content': cooking_user_prompt}
        ],
        temperature = 0
).choices[0].message.content

print(cooking_response)
```
